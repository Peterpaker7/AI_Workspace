```python
import pandas as pd
```

```python
dataset=pd.read_csv("50_Startups.csv")
```

```python
dataset=pd.get_dummies(dataset,drop_first=True).astype(int)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
```

```python
independent
```

```python
dependent=dataset[["Profit"]]
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test=tts(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion="squared_error",splitter='random',max_depth=10)
regressor=regressor.fit(x_train,y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree 
tree.plot_tree(regressor)
plt.show()
```

```python
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```
