```python
import pandas as pd
```

```python
dataset=pd.read_csv("CKD.csv",header=0)
```

```python
dataset=pd.get_dummies(dataset,drop_first=True).astype(int)
```

```python
dataset
```

```python
dataset.columns
```

```python
dataset.dtypes
```

```python
indep=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
```

```python
dep=dataset[['classification_yes']]
```

```python
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test=tts(indep,dep,test_size=1/3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import GridSearchCV
param_grid={'n_estimators':[200,500,1000],'criterion':["entropy", "log_loss"],}
grid=GridSearchCV(RandomForestClassifier(),param_grid,refit=True,verbose=3,cv=5,n_jobs=-1,scoring="f1_weighted")
grid.fit(x_train,y_train)
```

```python
re=grid.cv_results_
print("Best Parameter : ",grid.best_params_)
```

```python
grid_prediction=grid.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,grid_prediction)
print(cm)
```

```python
from sklearn.metrics import classification_report as clf_report
clf=clf_report(y_test,grid_prediction)
```

```python
print(clf)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,grid.predict_proba(x_test)[:,1])
```
