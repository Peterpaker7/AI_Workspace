```python
import pandas as pd
```

```python
dataset=pd.read_csv("Social_Network_Ads.csv")
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset.columns
```

```python
dataset=dataset.drop('User ID',axis=1)
```

```python
dataset.columns
```

```python
indep=dataset[['Age', 'EstimatedSalary','Gender_Male']]
```

```python
dep=dataset[['Purchased']]
```

```python
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test=tts(indep,dep,test_size=1/3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler as SC
sc=SC()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
param_grid={"solver":["newton-cg",'lbfgs','liblinear','saga'],'penalty':['l2']}
grid=GridSearchCV(LogisticRegression(),param_grid,refit=True,verbose=3,n_jobs=-1,scoring='f1_weighted')
```

```python
grid.fit(x_train,y_train)
```

```python
re=grid.cv_results_
```

```python
grid_prediction=grid.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,grid_prediction)
```

```python
from sklearn.metrics import classification_report as clf
clf_report=clf(y_test,grid_prediction)
```

```python
print(clf_report)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,grid.predict_proba(x_test)[:,1])
```
