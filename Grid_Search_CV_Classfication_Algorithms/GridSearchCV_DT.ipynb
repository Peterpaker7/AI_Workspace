```python
import pandas as pd
```

```python
dataset=pd.read_csv("Social_Network_Ads.csv")
```

```python
dataset=pd.get_dummies(dataset,drop_first=True).astype(int)
```

```python
dataset.columns
```

```python
dataset=dataset.drop('User ID',axis=1)
```

```python
dataset.columns
```

```python
inde=dataset[['Age', 'EstimatedSalary','Gender_Male']]
```

```python
dep=dataset[['Purchased']]
```

```python
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test=tts(inde,dep,test_size=1/3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import GridSearchCV
param_grid={'criterion':['entropy','log_loss'],'splitter':['best','random']}
grid=GridSearchCV(DecisionTreeClassifier(),param_grid,refit=True,verbose=3,n_jobs=-1,scoring="f1_weighted")
grid.fit(x_train,y_train)
```

```python
re=grid.cv_results_
```

```python
grid_prediction=grid.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,grid_prediction)
```

```python
print(cm)
```

```python
from sklearn.metrics import classification_report 
clf_report=classification_report(y_test,grid_prediction)
```

```python
print(clf_report)
```
