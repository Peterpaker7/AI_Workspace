```python
import pandas as pd
```

```python
dataset=pd.read_csv("50_Startups.csv")
```

```python
dataset=pd.get_dummies(dataset).astype(int)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_California', 'State_Florida', 'State_New York']]
```

```python
independent
```

```python
dependent=dataset[['Profit']]
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test=tts(independent,dependent,test_size=0.30,random_state=0)
```

```python
#from sklearn.preprocessing import StandardScaler as sc
#sc=sc()
#x_train=sc.fit_transform(x_train)
#x_test=sc.transform(x_test)
```

```python
y_test
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel="linear",C=1000)
regressor.fit(x_train,y_train)
```

```python
regressor.intercept_
```

```python
regressor.n_support_
```

```python
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```
